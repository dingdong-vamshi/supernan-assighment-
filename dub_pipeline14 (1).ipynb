```python
import whisper

print("loading whisper model...")
model = whisper.load_model("medium")

print("understanding speech and translating to english...")

result = model.transcribe("original.wav", task="translate")

english = result["text"]

print("\nenglish preview:\n")
print(english)

with open("english.txt","w",encoding="utf-8") as f:
    f.write(english)

print("\nenglish.txt created")
```

```python
from deep_translator import GoogleTranslator

with open("english.txt","r",encoding="utf-8") as f:
    eng = f.read()

hindi = GoogleTranslator(source="auto", target="hi").translate(eng)

print("\nhindi preview:\n")
print(hindi)

with open("hindi.txt","w",encoding="utf-8") as f:
    f.write(hindi)

print("\nhindi.txt created")
```

```python
import edge_tts
import asyncio

async def make_voice():
    with open("hindi.txt","r",encoding="utf-8") as f:
        text = f.read()

    communicate = edge_tts.Communicate(
        text=text,
        voice="hi-IN-SwaraNeural"
    )
    await communicate.save("hindi.wav")

await make_voice()

print("hindi voice created")
```
